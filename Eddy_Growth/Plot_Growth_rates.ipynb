```python
#Author: Giorgio M. Sarro (UChicago)
import numpy as np
import netCDF4 as nc
from scipy import interpolate
import matplotlib.pyplot as plt
import netCDF4 as nc
import scipy.io
from numpy import genfromtxt

#Open the data resulting from the growth rate calculation
files = nc.Dataset("/mnt/winds/data2/gmsarro/Rossbypalloza_project_22/LWA/run_LWA/eigenvalue_results_L0.5_U2.1.nc") 
k = files['k'][:]
largest_imaginary_eigenvalues = files['largest_imaginary_eigenvalues'][:]
files.close()

#Take the largest growth rates with a selected window of zonal wavenumber
window_size=35
largest_imaginary_eigenvalues_corrected= np.copy(largest_imaginary_eigenvalues)
for timing in range(int(len(largest_imaginary_eigenvalues)/10)):
   largest_imaginary_eigenvalues_corrected[timing*10:(timing+1)*10]= np.max(largest_imaginary_eigenvalues[timing*10:(timing+1)*10])

smoothed_array = np.convolve(largest_imaginary_eigenvalues_corrected, np.ones(window_size)/window_size, mode='same')

# Plot your data
plt.figure(figsize=(8, 6))
plt.plot(k, [x * k_value for x, k_value in zip(smoothed_array, k)], marker='o', linestyle='-', label='L=0.5')

# Add a title
plt.title('Growth Rates')

# Label the axes
plt.xlabel('k-values')
plt.ylabel('Growth Rate')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
