We assume perturbations to the PV field in the form:

$
\Psi'_1 = A_1(y) e^{i(kx-ct)} ;  
\Psi'_2 = A_2(y) e^{i(kx-ct)},
$

where $A_1(y)$ and $A_2(y)$ are the meridional structures in the upper and lower layer.
The imaginary component of $c$ is the growth rate


**Plot the growth rates:**

```python
#Author: Giorgio M. Sarro (UChicago)
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc

#Open the data resulting from the growth rate calculation
files = nc.Dataset("/mnt/winds/data2/gmsarro/Rossbypalloza_project_22/LWA/run_LWA/relaxation_eigenvalue_results_up.nc") 
k = files['k'][:]
largest_imaginary_eigenvalues = files['largest_imaginary_eigenvalues'][:]
files.close()


# Plot your data
plt.figure(figsize=(8, 6))
plt.plot(k, largest_imaginary_eigenvalues, marker='o', linestyle='-')
# Add a title
plt.title('Growth Rates')
# Label the axes
plt.xlabel('k-values')
plt.ylabel('Growth Rate')
plt.xlim(-0.1,2)
# Show the plot
plt.grid(True)
plt.show()
```

**We plot the spatial shape of $\Psi'_1$ and $\Psi'_2$ at peak growth here below.**


```python
#Author: Giorgio M. Sarro (UChicago)
import matplotlib.pyplot as plt
import numpy as np

#Get the x dimention from your original model
files = nc.Dataset("/mnt/winds/data2/gmsarro/Rossbypalloza_project_22/LWA/N128_0.0_2.0_0.1_1.0.3d.nc")
x= files['x'][:]
files.close()

#Open the growth calculations
files = nc.Dataset("/mnt/winds/data2/gmsarro/Rossbypalloza_project_22/LWA/run_LWA/relaxation_eigenvalue_results_up.nc")
y = files['y'][:]  
kk = files['k'][:]
growth = files['largest_imaginary_eigenvalues'][:]
peak_index = np.argmax(growth)
rk = kk[peak_index]

X, Y = np.meshgrid(x, y)

# Calculate the real and imaginary parts of the wave at the peak growth rate
wave_upper = np.outer(files['optimal_mode_upper_real'][:], np.cos(rk * x)) - np.outer(files['optimal_mode_upper_img'][:], np.sin(rk * x))
wave_lower = np.outer(files['optimal_mode_lower_real'][:], np.cos(rk * x)) - np.outer(files['optimal_mode_lower_img'][:], np.sin(rk * x))
files.close()




# Plotting
#set the limits
lev_up = np.arange(-.45,0.46, 0.1)

c1 = plt.contourf(X, Y, wave_upper, cmap='RdBu_r', levels=lev_up)
c3 = plt.contour(X, Y, wave_lower, colors='black', levels=lev_up, alpha=0.8)
plt.title('Perturbation Structure (lower layer black contours)')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-5,5)
plt.xlim(0,10)
plt.colorbar(c1)




plt.tight_layout()
plt.show()
```

We can observe a nice vertical tilt with height! This is exactly what we need for baroclinic instability.
